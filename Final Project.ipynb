## Median time_difference by year and by agency

On Cal OES's website, there is no clear data directory, but the California Hazardous Materials Spill / Release Notification Guidance said that “All significant spills or threatened releases of hazardous materials, including oil and radioactive materials, must be immediately reported. Notification shall be made by telephone.” https://www.caloes.ca.gov/wp-content/uploads/Fire-Rescue/Documents/CalOES-Spill_Booklet_Feb2014_FINAL_BW_Acc.pdf

Thus, I am going to check the time difference between the government's notified time and the actual incident time. 

### By year

```python
median_time_difference = hazmat_spills_complete[
    ['Incident Time', 'time_difference']
].groupby([pd.Grouper(key='Incident Time', axis=0, freq='A') ]).median()

median_time_difference
```

```python
median_time_difference = median_time_difference.reset_index()
median_time_difference['year'] = median_time_difference['Incident Time'].dt.year
```

```python
median_time_difference.columns = ['Incident Time', 'time_difference', 'year']
```

```python
median_time_difference
```

```python
median_time_difference = median_time_difference[['year', 'time_difference']].copy()
```

```python
median_time_difference
```

Since I encourtered the error of ValueError: Field "close_time" has type "timedelta64[ns]" which is not supported by Altair. Please convert to either a timestamp or a numerical value.

I need to convert `timedelta` to a different unit. 

```python
median_time_difference['time_difference_nanoseconds'] = median_time_difference['time_difference'].astype(int)
median_time_difference
```

```python
alt.Chart(median_time_difference[['year', 'time_difference_nanoseconds']]).mark_bar().encode(
    x='year:O',
    y='time_difference_nanoseconds',
).properties(
    title='California Spill Release Report: Median reporting time'
)
```

```python
alt.Chart(median_time_difference[['year', 'time_difference_nanoseconds']]).mark_bar().encode(
    x=alt.X('year:O', sort='-y'),
    y='time_difference_nanoseconds',
).properties(
    title='California Spill Release Report: Median reporting time'
)
```
